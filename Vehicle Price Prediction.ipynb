```python
final_df=pd.get_dummies(df,drop_first=True)
```

```python
final_df.head()
```

# Split of Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(final_df.drop(["Selling_Price"],axis=1),final_df["Selling_Price"],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

# Scaling the Year,Price,Kms_Driven

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit_transform(X_train.iloc[:,0:3])
X_train["Year"]=scaled[:,0]
X_train["Present_Price"]=scaled[:,1]
X_train["Kms_Driven"]=scaled[:,2]
```

```python
scaled=scaler.transform(X_test.iloc[:,0:3])
X_test["Year"]=scaled[:,0]
X_test["Present_Price"]=scaled[:,1]
X_test["Kms_Driven"]=scaled[:,2]
```

```python
X_train.head()
```

# Removing Outliers

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_train)
plt.scatter(X_pca[:,0],X_pca[:,1])
```

Internally checked the values and got 196,86.

```python
X_train.drop(196,inplace=True)
X_train.drop(86,inplace=True)
```

```python
y_train.drop([196,86],inplace=True)
```

# Random Forest Regressor

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
rf=RandomForestRegressor()
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_score_
```

```python
from sklearn import metrics
predictions=rf_random.predict(X_test)
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

# Support Vector Regressor

```python
from sklearn.svm import SVR
param={'C':[0.001,0.01,0.1,10,100],'kernel':['rbf','poly']}
rf_svc = RandomizedSearchCV(estimator = SVR(), param_distributions =param,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
rf_svc.fit(X_train,y_train)
```

```python
rf_svc.best_score_
```

```python
predictions=rf_svc.predict(X_test)
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('R-squared: {}'.format(metrics.r2_score(y_test,predictions)))
```
